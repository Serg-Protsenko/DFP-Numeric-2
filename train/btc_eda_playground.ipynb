```python
data_manager.plt_compare_features_ts(features_to_plt=['Price', 'Vol'], 
                                     titles=dict(fig_title="Price vs Vol",                                                 
                                                 xy=[1, .7],
                                                 xy_suptitle=[0.5,.95]
                                                ), 
                                     tot_figsize=(17,10),
                                     filters=["Date>='2021.02.01'", "Date<='2021.02.20'"], annotate=True)
```
