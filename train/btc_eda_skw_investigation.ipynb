## Power Transform 
(Yeo-J based, as we have zeros and negative values in data)

```python
df = data_manager.data_btc
df = df[[c for c in df.columns if c != "Price"]]

for i, col in enumerate(df.columns):
    print(f"{i} --> {col}")
```

```python
print(f'{"="*40} Power Transformed Features {"="*40}')
pd.DataFrame(power_transform(X=df)).hist(figsize=(15,9), bins=35);
```

# QuantileTransformer

```python
from sklearn.preprocessing import QuantileTransformer
qt = QuantileTransformer(n_quantiles=10, random_state=0, output_distribution='normal')

print(f'{"="*40} Quantie Transform Features {"="*40}')
pd.DataFrame(qt.fit_transform(X=df)).hist(figsize=(15,9), bins=35);
```
