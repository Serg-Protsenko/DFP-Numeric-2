```python
import pandas as pd
```

```python
files_prop = pd.read_csv('files_properties.csv').dropna()
files_prop
```

```python
from datetime import datetime
print("started: ", datetime.now())

for i, row in files_prop.iterrows():
    print("grabbing -->", row['FileName'], "...", end = " ")
    tmp_df = pd.read_csv(row['csv_url']).set_index("Date")
    tmp_df.to_csv(row['FileName'])
    print("done")

print(">>> Finished: ", datetime.now())
```
