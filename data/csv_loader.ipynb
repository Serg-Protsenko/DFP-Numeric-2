# Loading

```python
import os
import pandas as pd
from datetime import datetime
```

```python
files_prop = pd.read_csv('./csv_numeric_sources/files_properties.csv').dropna()
files_prop
```

```python
print("started: ", datetime.now(), '\n')

for i, row in files_prop.iterrows():
    print("grabbing -->", row['FileName'], "...", end = " ")
    tmp_df = pd.read_csv(row['csv_url']).set_index("Date")
    tmp_df.to_csv(os.path.join('csv_numeric_sources', row['FileName']))
    print("done")

print("\n>>> Finished: ", datetime.now())
```
